**Datos** **de** **Julio**

```python
df_caracteristicas_jul = pd.read_excel('JULIO/caracteristicas_julio.xlsx')
df_caracteristicas_jul.head(3)
```

```python
df_fuerza_jul = pd.read_excel('JULIO/fuerza_trabajo_julio.xlsx')
df_fuerza_jul.head(3)
```

```python
df_hogar_jul = pd.read_excel('JULIO/hogar_julio.xlsx')
df_hogar_jul.head(3)
```

```python
df_ocupados_jul = pd.read_excel('JULIO/ocupados_julio.xlsx')
df_ocupados_jul.head(3)
```

```python
df_caracteristicas_jul.columns
df_hogar_jul.columns
df_fuerza_jul.columns
df_ocupados_jul.columns
```

**Unión base de datos**

```python
base_julio = pd.merge(df_caracteristicas_jul, df_hogar_jul, on=["DIRECTORIO","SECUENCIA_P"],
    how="left")
```

```python
base_julio = pd.merge(base_julio, df_fuerza_jul, on=["DIRECTORIO","SECUENCIA_P","ORDEN"],
    how="left")
```

```python
df_ocupados_jul = df_ocupados_jul.drop(columns=["PERIODO","MES","PER","HOGAR","REGIS","AREA","CLASE"],
    errors="ignore")
```

```python
df_ocupados_jul = df_ocupados_jul.drop(columns=["FEX_C18","DPTO","FT"],
    errors="ignore")
```

```python
base_julio = pd.merge(base_julio, df_ocupados_jul, on=["DIRECTORIO","SECUENCIA_P","ORDEN"],
    how="left")
```

```python
base_julio.shape
base_julio.head()
```

**Archivo Divipola**

```python
divipola = pd.read_excel( '/content/drive/MyDrive/dataset_GEIH/2025-01-14 DIVIPOLA.xlsx',
    sheet_name='Departamentos', skiprows=8)
```
